```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MyApp") \
    .getOrCreate()
```

```python
data = [
    (1, "Rahul Sharma", "Hyderabad", "Electronics", "Laptop", 1, 65000, "Delivered"),
    (2, "Priya Reddy", "Bangalore", "Electronics", "Mobile", 2, 25000, "Delivered"),
    (3, "Amit Kumar", "Mumbai", "Furniture", "Chair", 4, 3500, "Pending"),
    (4, "Sneha Patel", "Chennai", "Electronics", "Tablet", 1, 30000, "Cancelled"),
    (5, "Farhan Ali", "Delhi", "Furniture", "Table", 2, 12000, "Delivered"),
    (6, "Neha Singh", "Hyderabad", "Fashion", "Shoes", 3, 2500, "Delivered"),
    (7, "Arjun Verma", "Pune", "Fashion", "Watch", 1, 8000, "Pending"),
    (8, "Meera Nair", "Bangalore", "Electronics", "TV", 1, 45000, "Delivered"),
    (9, "Kiran Rao", "Hyderabad", "Furniture", "Desk", 1, 15000, "Delivered"),
    (10, "Ravi Kumar", "Mumbai", "Electronics", "Mobile", 1, 28000, "Delivered")
]

columns = [
    "OrderID",
    "CustomerName",
    "City",
    "Category",
    "Product",
    "Quantity",
    "Price",
    "OrderStatus"
]
```

```python
df = df.withColumn("total_amount", df.Quantity * df.Price)
df.show()
```

```python
df = df.withColumn("discount", df.total_amount * 0.10)
df.show()
```

```python
df = df.withColumnRenamed("CustomerName", "customer")
```

```python
df = df.withColumnRenamed("OrderStatus", "status")
```

```python
from pyspark.sql import functions as F

df = df.withColumn("total_amount", F.col("Quantity") * F.col("Price"))
df = df.withColumn("discount", F.col("total_amount") * 0.10)
```

```python
df.select(F.sum("total_amount").alias("Total_Revenue")).show()
```

```python
df.select(F.avg("total_amount").alias("Average_Order_Value")).show()
```

```python
df.select(F.max("total_amount").alias("Highest_Order_Value")).show()
```

```python
df.select(F.min("total_amount").alias("Lowest_Order_Value")).show()
```

```python
from pyspark.sql.functions import when
df = df.withColumn(
    "order_value_category",
    when(F.col("total_amount") >= 50000, "High")
    .when(F.col("total_amount") >= 20000, "Medium")
    .otherwise("Low")
)
```

```python
df.groupBy("order_value_category").count().show()
```

```python
from pyspark.sql import functions as F

df = df.withColumn(
    "final_amount",
    F.col("total_amount") - F.col("discount")
)
```

```python
df.groupBy("order_value_category").count().show()
```

```python
df.filter(F.col("status") == "Delivered") \
  .select("OrderID", "customer", "final_amount") \
  .show()
```

```python
df.filter(F.col("status") == "Delivered") \
  .groupBy("City") \
  .agg(F.sum("final_amount").alias("Delivered_Revenue")) \
  .show()
```
