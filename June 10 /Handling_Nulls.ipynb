```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql import functions as F

spark = SparkSession.builder.appName("NullHandling").getOrCreate()

df = spark.read.csv("bookings.csv", header=True, inferSchema=True)
```

```python
df.dropna(subset=["booking_amount"]).show()
```

```python
df.dropna(
    subset=["customer_name","service_type","booking_amount"]
).show()
```

```python
df = df.fillna({"city":"Unknown"})
```

```python
df = df.fillna({"provider":"Not Available"})
```

```python
df = df.fillna({"payment_mode":"Not Provided"})
```

```python
df = df.fillna({"booking_status":"Unknown"})
```

```python
df = df.fillna({"booking_amount":0})
```

```python
df = df.withColumn(
    "data_quality_status",
    when(
        col("city").isNull() |
        col("service_type").isNull() |
        col("provider").isNull() |
        col("booking_amount").isNull(),
        "Incomplete"
    ).otherwise("Complete")
)
```

```python
df.groupBy("data_quality_status").count().show()
```

```python
df.filter(col("data_quality_status")=="Complete").show()
```

```python
df.filter(col("data_quality_status")=="Incomplete").show()
```

```python
df = df.withColumn(
    "tax",
    col("booking_amount") * 0.05
)
```

```python
df = df.withColumn(
    "final_amount",
    col("booking_amount") + col("tax")
)
```

```python
df.filter(col("booking_status")=="Confirmed") \
  .agg(sum("booking_amount").alias("Revenue")) \
  .show()
```

```python
df.fillna({"service_type":"Unknown"}) \
  .groupBy("service_type") \
  .count() \
  .show()
```

```python
df.fillna({"city":"Unknown"}) \
  .groupBy("city") \
  .count() \
  .show()
```

```python
df.fillna({"booking_amount":0}) \
  .agg(avg("booking_amount").alias("Average")) \
  .show()
```

```python
df.dropna(subset=["booking_amount"]) \
  .agg(avg("booking_amount").alias("Average")) \
  .show()
```

```python
avg1 = df.fillna({"booking_amount":0}) \
         .agg(avg("booking_amount")) \
         .collect()[0][0]

avg2 = df.dropna(subset=["booking_amount"]) \
         .agg(avg("booking_amount")) \
         .collect()[0][0]

print("Average after replacing null with 0 =", avg1)
print("Average after dropping null rows =", avg2)
```

```python
df.write.mode("overwrite").parquet("clean_bookings.parquet")
```
