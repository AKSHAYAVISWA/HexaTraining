```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, datediff, when, count

spark = SparkSession.builder \
    .appName("Week3_Order_Analysis") \
    .getOrCreate()

print("Spark Started Successfully!")
```

```python
import pandas as pd

orders = pd.DataFrame({
    "order_id":[101,102,103,104,105,106,107,108],
    "customer_id":[1,2,3,4,5,2,1,3],
    "order_date":[
        "2024-01-01",
        "2024-01-02",
        "2024-01-03",
        "2024-01-04",
        "2024-01-05",
        "2024-01-06",
        "2024-01-07",
        "2024-01-08"
    ],
    "delivery_date":[
        "2024-01-05",
        "2024-01-10",
        "2024-01-04",
        "2024-01-12",
        "2024-01-07",
        "2024-01-15",
        "2024-01-08",
        "2024-01-18"
    ]
})

orders.to_csv("orders.csv", index=False)

print("orders.csv created")
orders
```

```python
customers = pd.DataFrame({
    "customer_id":[1,2,3,4,5],
    "customer_name":[
        "Akshaya",
        "Rahul",
        "Priya",
        "Arun",
        "Sneha"
    ],
    "region":[
        "South",
        "North",
        "East",
        "West",
        "South"
    ]
})

customers.to_csv("customers.csv", index=False)

print("customers.csv created")
customers
```

```python
orders_df = spark.read.csv(
    "orders.csv",
    header=True,
    inferSchema=True
)

customers_df = spark.read.csv(
    "customers.csv",
    header=True,
    inferSchema=True
)

print("Orders Table")
orders_df.show()

print("Customers Table")
customers_df.show()
```

```python
joined_df = orders_df.join(
    customers_df,
    on="customer_id",
    how="inner"
)

print("Joined Table")
joined_df.show()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Week3").getOrCreate()

orders = spark.createDataFrame(orders)
customers = spark.createDataFrame(customers)
```

```python

joined = orders.join(customers, on="customer_id", how="inner")

joined.show()
```

```python
joined_df = joined_df.withColumn(
    "delivery_days",
    datediff(col("delivery_date"), col("order_date"))
)

joined_df = joined_df.withColumn(
    "Delayed",
    when(col("delivery_days") > 5, "Yes").otherwise("No")
)

print("Orders with Delay Status")
joined_df.show()
```

```python
result = joined_df.filter(col("Delayed")=="Yes") \
                  .groupBy("region") \
                  .agg(count("*").alias("Delayed_Orders"))

print("Delayed Orders by Region")
result.show()
```

```python
result.coalesce(1).write \
      .mode("overwrite") \
      .option("header", True) \
      .csv("output")

print("Output saved successfully!")
```
