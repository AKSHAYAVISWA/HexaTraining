```python
freq = rdd.countByValue()

for activity,count in sorted(
    freq.items(),
    key=lambda x:x[1],
    reverse=True
):
    print(activity,count)
```

```python
freq_rdd = rdd.map(
    lambda x:(x,1)
).reduceByKey(
    lambda a,b:a+b
)

freq_rdd.saveAsTextFile(
    "activity_frequency"
)
```

```python
freq = rdd.countByValue()

print("Total Activities :", rdd.count())

print("Unique Activities :",
      rdd.distinct().count())

print("Most Common :",
      max(freq.items(),
          key=lambda x:x[1]))

print("Least Common :",
      min(freq.items(),
          key=lambda x:x[1]))
```
