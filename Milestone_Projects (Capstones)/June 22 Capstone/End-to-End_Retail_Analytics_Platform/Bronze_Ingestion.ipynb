```python
from pyspark.sql import Row

preferences_data = [
    Row(
        customer_id="C101",
        preferred_channel="Online",
        loyalty=Row(tier="Gold", points=1200)
    ),
    Row(
        customer_id="C102",
        preferred_channel="Store",
        loyalty=Row(tier="Silver", points=700)
    ),
    Row(
        customer_id="C104",
        preferred_channel="Online",
        loyalty=Row(tier="Platinum", points=2200)
    ),
    Row(
        customer_id="C108",
        preferred_channel="Mobile App",
        loyalty=Row(tier="Gold", points=1500)
    )
]

preferences_df = spark.createDataFrame(preferences_data)

display(preferences_df)
preferences_df.printSchema()
```

```python
preferences_df.coalesce(1).write \
.mode("overwrite") \
.json("dbfs:/FileStore/retail/customer_preferences.json")
```
