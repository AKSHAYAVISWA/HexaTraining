```python
final_df = spark.table("silver_healthcare")

display(final_df)
```

```python
final_df.createOrReplaceTempView("healthcare")
```

```python
%sql
SELECT SUM(final_bill) AS total_revenue
FROM healthcare;
```

```python
%sql
SELECT department,
SUM(final_bill) AS revenue
FROM healthcare
GROUP BY department
ORDER BY revenue DESC;
```

```python
%sql
SELECT city,
SUM(final_bill) AS revenue
FROM healthcare
GROUP BY city
ORDER BY revenue DESC;
```

```python
%sql
SELECT *
FROM healthcare
WHERE status='Completed';
```

```python
%sql
SELECT patient_name,
SUM(final_bill) total_bill
FROM healthcare
GROUP BY patient_name
ORDER BY total_bill DESC;
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *
```

```python
doctor_rank = final_df.groupBy(
    "doctor_name"
).agg(
    sum("final_bill").alias("revenue")
)

doctor_rank = doctor_rank.withColumn(
    "rank",
    rank().over(Window.orderBy(desc("revenue")))
)

display(doctor_rank)
```

```python
dept_rank = final_df.groupBy(
    "department"
).agg(
    sum("final_bill").alias("revenue")
)

dept_rank = dept_rank.withColumn(
    "rank",
    rank().over(Window.orderBy(desc("revenue")))
)

display(dept_rank)
```

```python
top3 = final_df.groupBy(
    "patient_name"
).agg(
    sum("final_bill").alias("billing")
)

top3 = top3.withColumn(
    "rank",
    rank().over(Window.orderBy(desc("billing")))
).filter("rank <= 3")

display(top3)
```

```python
doctor_dept = final_df.groupBy(
    "department",
    "doctor_name"
).agg(
    sum("final_bill").alias("revenue")
)

top_doc = doctor_dept.withColumn(
    "rank",
    rank().over(
        Window.partitionBy("department")
        .orderBy(desc("revenue"))
    )
).filter("rank=1")

display(top_doc)
```

```python
daily_revenue = final_df.groupBy(
    "appointment_date"
).agg(
    sum("final_bill").alias("daily_revenue")
)

running_revenue = daily_revenue.withColumn(
    "running_revenue",
    sum("daily_revenue").over(
        Window.orderBy("appointment_date")
    )
)

display(running_revenue)
```

```python
%sql
CREATE OR REPLACE TEMP VIEW patient_updates AS
SELECT
'P101' AS patient_id,
'Chennai' AS city,
'Inactive' AS insurance_status;
```

```python
%sql
MERGE INTO bronze_patients p
USING patient_updates u
ON p.patient_id=u.patient_id

WHEN MATCHED THEN
UPDATE SET
p.city=u.city,
p.insurance_status=u.insurance_status;
```
