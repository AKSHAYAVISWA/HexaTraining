```python
patients_df = spark.read.option("header",True).csv(
"/Volumes/databricks_akshaya/default/healthcare/patients.csv"
)

display(patients_df)
```

```python
doctors_df = spark.read.option("header",True).csv(
"/Volumes/databricks_akshaya/default/healthcare/doctors.csv"
)

display(doctors_df)
```

```python
appointments_df = spark.read.option("header",True).csv(
"/Volumes/databricks_akshaya/default/healthcare/appointments.csv"
)

display(appointments_df)
```

```python
preferences_df = spark.read \
.option("multiline","true") \
.json("/Volumes/databricks_akshaya/default/healthcare/patient_preferences.json")

display(preferences_df)
```

```python
patients_df.printSchema()
doctors_df.printSchema()
appointments_df.printSchema()
preferences_df.printSchema()
```

```python
patients_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze_patients")

doctors_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze_doctors")

appointments_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze_appointments")

preferences_df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("bronze_preferences")
```
