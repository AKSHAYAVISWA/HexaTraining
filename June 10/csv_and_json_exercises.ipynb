```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
  .appName("Working with files")\
  .getOrCreate()
```

```python
df = spark.read.csv(
    "flight_bookings.csv",
    header=True,
    inferSchema=True
)
df.show()
```

```python
from pyspark.sql import functions as F
```

```python
df.show()
```

```python
df.select("passenger_name","from_city","to_city").show()
```

```python
df.filter(F.col("status")=="Confirmed").show()
```

```python
df.filter(F.col("status") == "Cancelled").show()
```

```python
df.filter(F.col("status") == "Pending").show()
```

```python
df.filter(F.col("ticket_price") > 20000).show()
```

```python
df.filter(
    F.col("to_city").isin("Dubai", "Singapore", "London")
).show()
```

```python
df.count()
```

```python
df.groupBy("airline").count().show()
```

```python
df.groupBy("status").count().show()
```

```python
df.filter(
    F.col("status") == "Confirmed"
).agg(
    F.sum("ticket_price").alias("total_revenue")
).show()
```

```python
df.groupBy("airline").agg(
    F.avg("ticket_price").alias("avg_price")
).show()
```

```python
df.agg(
    F.max("ticket_price").alias("highest_price")
).show()
```

```python
df.agg(
    F.min("ticket_price").alias("lowest_price")
).show()
```

```python
df.orderBy(
    F.col("ticket_price").desc()
).show()
```

```python
df.withColumn(
    "tax",
    F.col("ticket_price") * 0.05
).show()
```

```python
df.withColumn(
    "tax",
    F.col("ticket_price") * 0.05
).withColumn(
    "final_price",
    F.col("ticket_price") + F.col("tax")
).show()
```

```python
df.withColumn(
    "price_category",
    F.when(F.col("ticket_price") >= 30000, "Premium")
     .when(F.col("ticket_price") >= 10000, "Standard")
     .otherwise("Budget")
).show()
```

```python
df.withColumn(
    "price_category",
    F.when(F.col("ticket_price") >= 30000, "Premium")
     .when(F.col("ticket_price") >= 10000, "Standard")
     .otherwise("Budget")
).groupBy(
    "price_category"
).count().show()
```

```python
df.filter(
    F.col("status") == "Confirmed"
).write.mode("overwrite").parquet(
    "confirmed_flights.parquet"
)
```

```python
hotels_df = spark.read.option(
    "multiline",
    "true"
).json("hotels.json")

hotels_df.show(truncate=False)
hotels_df.printSchema()
```

```python
from pyspark.sql import functions as F
```

```python
hotels_df.show()
```

```python
hotels_df.select(
    "hotel_name",
    "city",
    "rating"
).show()
```

```python
hotels_df.filter(
    F.col("category") == "Luxury"
).show()
```

```python
hotels_df.filter(
    F.col("rating") > 4.5
).show()
```

```python
hotels_df.filter(
    F.col("rooms_available") > 15
).show()
```

```python
hotels_df.filter(
    F.col("price_per_night") > 10000
).show()
```

```python
hotels_df.filter(
    F.col("city").isin("Dubai", "London")
).show()
```

```python
hotels_df.filter(
    F.col("contact.phone").isNull()
).show()
```

```python
hotels_df.filter(
    F.col("contact.email").isNull()
).show()
```

```python
hotels_df.select(
    "hotel_name",
    "contact.phone",
    "contact.email"
).show()
```

```python
hotels_df.filter(
    F.array_contains(
        F.col("amenities"),
        "wifi"
    )
).show()
```

```python
hotels_df.filter(
    F.array_contains(
        F.col("amenities"),
        "spa"
    )
).show()
```

```python
hotels_df.groupBy(
    "city"
).count().show()
```

```python
hotels_df.groupBy(
    "category"
).count().show()
```

```python
hotels_df.groupBy(
    "category"
).agg(
    F.avg("rating").alias("avg_rating")
).show()
```

```python
hotels_df.groupBy(
    "city"
).agg(
    F.avg("price_per_night").alias("avg_price")
).show()
```

```python
hotels_df.agg(
    F.max("price_per_night").alias("highest_price")
).show()
```

```python
hotels_df.withColumn(
    "total_potential_revenue",
    F.col("rooms_available")
    * F.col("price_per_night")
).show()
```

```python
hotels_df.orderBy(
    F.col("rating").desc()
).show()
```

```python
flattened_df = hotels_df.select(
    "hotel_id",
    "hotel_name",
    "city",
    "category",
    "rating",
    "rooms_available",
    "price_per_night",
    "contact.phone",
    "contact.email"
)

flattened_df.write.mode(
    "overwrite"
).parquet(
    "hotels_flattened.parquet"
)
```
