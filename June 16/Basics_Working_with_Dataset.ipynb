```python
w_revenue = Window.orderBy(F.desc("total_amount"))
```

```python
df_calc.withColumn("row_num", F.row_number().over(w_revenue)).show()
```

```python
df_calc.withColumn("rank", F.rank().over(w_revenue)).show()
```

```python
df_calc.withColumn("dense_rank", F.dense_rank().over(w_revenue)).show()
```

```python
df_calc.withColumn("row_num", F.row_number().over(w_revenue)) \
       .filter(F.col("row_num") <= 5).show()
```

```python
df_calc.withColumn("row_num", F.row_number().over(w_revenue)) \
       .filter(F.col("row_num") == 1).show()
```

```python
w_city = Window.partitionBy("city").orderBy(F.desc("total_amount"))
df_calc.withColumn("row_num", F.row_number().over(w_city)) \
       .filter(F.col("row_num") == 1).show()
```

```python
w_category = Window.partitionBy("category").orderBy(F.desc("total_amount"))
df_calc.withColumn("row_num", F.row_number().over(w_category)) \
       .filter(F.col("row_num") == 1).show()
```

```python
w_running = Window.orderBy("order_id").rowsBetween(Window.unboundedPreceding, 0)
df_calc.withColumn("running_total", F.sum("total_amount").over(w_running)).show()
```

```python
w_order = Window.orderBy("order_id")
df_calc.withColumn("prev_revenue", F.lag("total_amount").over(w_order)).show()
```

```python
df_calc.withColumn("next_revenue", F.lead("total_amount").over(w_order)).show()
```

```python
df_calc.withColumn("prev_revenue", F.lag("total_amount").over(w_order)) \
       .withColumn(
           "comparison",
           F.when(F.col("total_amount") > F.col("prev_revenue"), "Higher")
            .when(F.col("total_amount") < F.col("prev_revenue"), "Lower")
            .otherwise("Same/First"),
       ).show()
```

```python
df_calc.withColumn("prev_revenue", F.lag("total_amount").over(w_order)) \
       .withColumn("revenue_diff", F.col("total_amount") - F.col("prev_revenue")) \
       .show()
```

```python
df_calc.withColumn("row_num", F.row_number().over(w_city)) \
       .filter(F.col("row_num") <= 2).show()
```

```python
df_calc.withColumn("row_num", F.row_number().over(w_category)) \
       .filter(F.col("row_num") <= 3).show()
```

```python
revenue_ranking_report = (
    df_calc.withColumn("row_num", F.row_number().over(w_revenue))
           .withColumn("rank", F.rank().over(w_revenue))
           .withColumn("dense_rank", F.dense_rank().over(w_revenue))
           .select(
               "order_id", "product", "city", "category",
               "total_amount", "row_num", "rank", "dense_rank",
           )
)
revenue_ranking_report.show()
```

```python
source_df = spark.createDataFrame(sales_data, columns)

# 92. Create calculated columns.
etl_df = (
    source_df
    .withColumn("total_amount", F.col("quantity") * F.col("price"))
    .withColumn("tax", F.col("total_amount") * 0.05)
    .withColumn("final_amount", F.col("total_amount") + F.col("tax"))
)
```

```python
etl_completed = etl_df.filter(F.col("status") == "Completed")
```

```python
etl_completed = etl_completed.withColumn("revenue", F.col("total_amount"))
```

```python
city_revenue_report = etl_completed.groupBy("city").agg(
    F.sum("revenue").alias("total_revenue")
).orderBy(F.desc("total_revenue"))
city_revenue_report.show()
```

```python
category_revenue_report = etl_completed.groupBy("category").agg(
    F.sum("revenue").alias("total_revenue")
).orderBy(F.desc("total_revenue"))
category_revenue_report.show()
```

```python
etl_completed.write.mode("overwrite").parquet("/tmp/sales_output_parquet")
```

```python
parquet_df = spark.read.parquet("/tmp/sales_output_parquet")
parquet_df.show()
```

```python
parquet_df.createOrReplaceTempView("sales_parquet")
```

```python
final_business_report = spark.sql("""
    SELECT
        city,
        category,
        COUNT(*) AS total_orders,
        SUM(quantity) AS total_quantity,
        SUM(revenue) AS total_revenue,
        ROUND(AVG(revenue), 2) AS avg_order_revenue
    FROM sales_parquet
    GROUP BY city, category
    ORDER BY total_revenue DESC
""")
final_business_report.show()
```
