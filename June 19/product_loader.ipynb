```python
products = [
    (1, "Laptop", 50000),
    (2, "Mobile", 20000),
    (3, "Headphones", 3000),
    (4, "Monitor", 15000)
]

df = spark.createDataFrame(
    products,
    ["product_id", "product_name", "price"]
)

display(df)

df.createOrReplaceTempView("products")
```

```python
%sql
SELECT AVG(price) AS avg_price
FROM products;
```
