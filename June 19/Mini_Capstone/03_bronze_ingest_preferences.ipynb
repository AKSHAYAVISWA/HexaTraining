```python
preferences_file = "/Volumes/ak_databricks/default/airline_files/passenger_preferences.json"

preferences_df = spark.read.option(
    "multiLine","true"
).json(preferences_file)

preferences_flat = preferences_df.select(
    "passenger_name",
    "meal",
    "seat",
    "extra_baggage"
)

display(preferences_flat)
```

```python

preferences_flat.write.mode("overwrite").saveAsTable("bronze_preferences")
```
