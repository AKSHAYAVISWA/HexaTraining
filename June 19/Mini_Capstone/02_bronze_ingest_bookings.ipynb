```python
bookings_file = "/Volumes/ak_databricks/default/airline_files/bookings.csv - Sheet1.csv"

from pyspark.sql.functions import split,col,regexp_replace

raw_bookings = spark.read.text(bookings_file)

clean_bookings = raw_bookings.select(
    regexp_replace(col("value"), '"', '').alias("value")
)

bookings_df = clean_bookings.select(
    split(col("value"), ",").getItem(0).alias("booking_id"),
    split(col("value"), ",").getItem(1).alias("flight_id"),
    split(col("value"), ",").getItem(2).alias("passenger_name"),
    split(col("value"), ",").getItem(3).alias("travel_class"),
    split(col("value"), ",").getItem(4).cast("double").alias("ticket_price"),
    split(col("value"), ",").getItem(5).alias("booking_date")
)

bookings_df = bookings_df.filter(
    col("booking_id") != "booking_id"
)

display(bookings_df)
```

```python
bookings_df.write.mode("overwrite").saveAsTable("bronze_bookings")
```
