```python
flights_file = "/Volumes/ak_databricks/default/airline_files/Flights.csv"

from pyspark.sql.functions import split,col

raw_flights = spark.read.text(flights_file)

flights_df = raw_flights.select(
    split(col("value"), ",").getItem(0).alias("flight_id"),
    split(col("value"), ",").getItem(1).alias("airline"),
    split(col("value"), ",").getItem(2).alias("from_city"),
    split(col("value"), ",").getItem(3).alias("to_city"),
    split(col("value"), ",").getItem(4).alias("duration"),
    split(col("value"), ",").getItem(5).alias("status")
)

flights_df = flights_df.filter(col("flight_id") != "flight_id")

display(flights_df)
flights_df.printSchema()
```

```python
flights_df.write.mode("overwrite").saveAsTable("bronze_flights")
```
