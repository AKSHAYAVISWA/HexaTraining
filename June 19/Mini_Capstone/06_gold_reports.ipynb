```python
journey_df = spark.table("silver_passenger_journey")

flights_df = spark.table("bronze_flights")
```

```python
airline_report = journey_df.groupBy(
    "airline"
).sum("revenue")
display(airline_report)
```

```python
route_report = journey_df.groupBy(
    "from_city",
    "to_city"
).sum("revenue")

display(route_report)
```

```python
pref_report = journey_df.select(
    "passenger_name",
    "meal",
    "seat"
)

display(pref_report)
```

```python
delay_report = flights_df.select(
    "flight_id",
    "status"
)

display(delay_report)
```
