```python
customers = [
    (1, "Akshaya", "Chennai"),
    (2, "Sam", "Mumbai"),
    (3, "Priya", "Chennai"),
    (4, "John", "Delhi")
]

df = spark.createDataFrame(
    customers,
    ["cust_id", "name", "city"]
)

df.write.mode("overwrite").saveAsTable("customer_table")

print("Customer Table Created")
```
